**DistribuciÃ³n de evasiÃ³n**

comprender cÃ³mo estÃ¡ distribuida la variable "churn" (evasiÃ³n) entre los clientes. Utiliza grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Contar valores 0 y 1
churn_counts = df_normalizado['abandono'].value_counts()

# Mapear los valores 0 y 1 a nombres legibles
label_map = {0: 'Permanecen', 1: 'Dieron de Baja'}
labels = [label_map.get(i, str(i)) for i in churn_counts.index]

# GrÃ¡fico de pastel
plt.figure(figsize=(6, 6))
plt.pie(
    churn_counts,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66c2a5', '#fc8d62']
)
plt.title('ProporciÃ³n de Clientes: Permanecen vs Dieron de Baja')
plt.axis('equal')
plt.show()
```

**Recuento de evasiÃ³n por variables categÃ³ricas**

Exploraremos cÃ³mo se distribuye la evasiÃ³n segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras.

Este anÃ¡lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudarÃ¡ a orientar acciones estratÃ©gicas.

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df_normalizado, x='genero', hue='abandono', palette='Set2')

plt.title("Porcentaje de EvasiÃ³n por GÃ©nero", fontsize=16)
plt.xlabel("GÃ©nero", fontsize=12)
plt.ylabel("Cantidad de Clientes", fontsize=12)
plt.legend(title='Abandono', fontsize=10, title_fontsize=12)

# Crear tabla resumen para valores y porcentajes
tabla = df_normalizado.groupby(['genero', 'abandono']).size().reset_index(name='count')
total_por_genero = df_normalizado.groupby('genero').size().reset_index(name='total')

# Diccionario para acceder total por gÃ©nero
total_dict = dict(zip(total_por_genero['genero'], total_por_genero['total']))

# Los grupos en el eje x (gÃ©neros)
generos = df_normalizado['genero'].unique()
# Los niveles de hue ('SÃ­', 'No')
abandono_niveles = df_normalizado['abandono'].unique()

# Para cada barra, encontrar la posiciÃ³n y poner texto
for i, genero in enumerate(generos):
    for j, abandono in enumerate(abandono_niveles):
        # Buscar el conteo para ese grupo
        fila = tabla[(tabla['genero'] == genero) & (tabla['abandono'] == abandono)]
        if not fila.empty:
            count = int(fila['count'])
            total = total_dict[genero]
            porcentaje = count / total * 100

            # Calcular posiciÃ³n de la barra en el grÃ¡fico:
            # El ancho de cada barra es aproximadamente 0.8 / nÃºmero de hues (a ojo)
            n_hues = len(abandono_niveles)
            bar_width = 0.8 / n_hues
            # PosiciÃ³n x base del grupo i en el eje x
            x = i - 0.4 + bar_width/2 + j*bar_width

            # Altura de la barra
            altura = count

            ax.text(x, altura + max(tabla['count']) * 0.02, f'{count}\n({porcentaje:.1f}%)',
                    ha='center', va='bottom', fontsize=10)

plt.ylim(0, tabla['count'].max() * 1.15)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Abandono: SÃ­
abandono = df_normalizado[df_normalizado['abandono'] == 'SÃ­']['tipo_contrato'].value_counts()
axes[0].pie(
    abandono,
    labels=abandono.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2"),
    wedgeprops={'width': 0.4}
)
axes[0].set_title("Clientes que Abandonaron")

# Abandono: No
no_abandono = df_normalizado[df_normalizado['abandono'] == 'No']['tipo_contrato'].value_counts()
axes[1].pie(
    no_abandono,
    labels=no_abandono.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2"),
    wedgeprops={'width': 0.4}
)
axes[1].set_title("Clientes que No Abandonaron")

plt.suptitle("Comparativa de EvasiÃ³n por Tipo de Contrato", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Crear tabla de contingencia con conteos
tabla = pd.crosstab(df_normalizado['metodo_pago'], df_normalizado['abandono'])

# Calcular proporciones (opcional para que sea proporcional)
tabla_prop = tabla.div(tabla.sum(axis=1), axis=0)

# Colores para abandono "No" y "SÃ­"
colores = ['#66c2a5', '#fc8d62']

fig, ax = plt.subplots(figsize=(10,6))

# Graficar barras apiladas proporcionales
tabla_prop.plot(kind='bar', stacked=True, color=colores, ax=ax)

ax.set_title('ProporciÃ³n de EvasiÃ³n por MÃ©todo de Pago', fontsize=16)
ax.set_ylabel('ProporciÃ³n de Clientes')
ax.set_xlabel('MÃ©todo de Pago')
ax.legend(title='Abandono', fontsize=12, title_fontsize=14)
plt.xticks(rotation=45)

# Agregar etiquetas con porcentaje encima de cada secciÃ³n
for i, metodo in enumerate(tabla_prop.index):
    bottom = 0
    for abandono in tabla_prop.columns:
        altura = tabla_prop.loc[metodo, abandono]
        if altura > 0.05:  # Mostrar solo si es >5%
            ax.text(i, bottom + altura/2, f'{altura:.0%}', ha='center', va='center', color='white', fontsize=11)
        bottom += altura

plt.tight_layout()
plt.show()
```
